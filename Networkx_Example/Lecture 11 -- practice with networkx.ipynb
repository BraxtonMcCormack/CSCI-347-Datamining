```python
# Creates a graph from the txt file
G_fb = nx.read_edgelist("facebook_combined.txt", create_using = nx.Graph(), nodetype=int)
```

```python
# G_fb.info() no longer works :(
print(G_fb)
```

```python
pos = nx.spring_layout(G_fb) # this takes a while to run
```

```python
bc_fb = nx.betweenness_centrality(G_fb, normalized=True, endpoints=True)
```

```python
node_color = [20000*G_fb.degree(v) for v in G_fb] #if a node degree is 1, itll become 20,000 and
node_size = [v*1000 for v in bc_fb.values()] #1000 adds scalar for visibility
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(20,20))
nx.draw_networkx(G_fb, pos=pos, with_labels=False, node_color=node_color, node_size=node_size)
```
